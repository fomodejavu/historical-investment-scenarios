# 💰 Dividend Reinvestment (DRIP) Analysis

This notebook illustrates the compounding power of dividend reinvestment using scenarios from this dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json

# DRIP vs No-DRIP comparison function
def simulate_drip(
    initial_investment: float,
    annual_price_return: float,
    annual_dividend_yield: float,
    years: int,
    drip: bool = True
) -> list:
    """Simulate annual portfolio value with or without DRIP."""
    portfolio_value = initial_investment
    values = [initial_investment]
    
    for year in range(years):
        dividend = portfolio_value * annual_dividend_yield
        price_gain = portfolio_value * annual_price_return
        
        if drip:
            portfolio_value += price_gain + dividend
        else:
            portfolio_value += price_gain  # Dividends taken as cash
        
        values.append(portfolio_value)
    
    return values
```

```python
# Simulate JNJ-like scenario: 8% price return + 2.5% dividend yield over 21 years
years = 21
initial = 10000
price_return = 0.082
div_yield = 0.025

drip_values = simulate_drip(initial, price_return, div_yield, years, drip=True)
no_drip_values = simulate_drip(initial, price_return, div_yield, years, drip=False)

year_labels = list(range(2004, 2004 + years + 1))

fig, ax = plt.subplots(figsize=(12, 7))

ax.fill_between(year_labels, drip_values, no_drip_values,
                alpha=0.15, color='#2563eb', label='DRIP advantage')
ax.plot(year_labels, drip_values, color='#2563eb', linewidth=2.5,
        label=f'With DRIP: ${drip_values[-1]:,.0f}')
ax.plot(year_labels, no_drip_values, color='#6b7280', linewidth=2,
        linestyle='--', label=f'Price only (no DRIP): ${no_drip_values[-1]:,.0f}')
ax.axhline(initial, color='red', linestyle=':', alpha=0.5, label=f'Initial: ${initial:,}')

drip_advantage = drip_values[-1] - no_drip_values[-1]

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Portfolio Value (USD)', fontsize=12)
ax.set_title(f'DRIP vs No-DRIP: $10,000 Investment\n({price_return*100:.1f}% price return + {div_yield*100:.1f}% dividend yield over {years} years)',
             fontsize=13)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1000:.0f}K'))
ax.legend(fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print(f"DRIP advantage after {years} years: ${drip_advantage:,.0f}")
print(f"DRIP contributed {drip_advantage/drip_values[-1]*100:.1f}% of total final value")
```

```python
# Load real dividend scenario data
with open('../scenarios/dividends/jnj-drip-2004-10000usd.json') as f:
    jnj = json.load(f)

print("=== Johnson & Johnson DRIP Scenario ===")
print(f"Initial investment: ${jnj['initial_investment_usd']:,}")
print(f"Value today (with DRIP): ${jnj['value_today_usd']:,}")
print(f"Value without DRIP: ${jnj['drip_breakdown']['price_appreciation_only_value']:,}")
print(f"DRIP added value: ${jnj['drip_breakdown']['drip_added_value']:,}")
print(f"DRIP contribution: {jnj['drip_breakdown']['drip_contribution_pct']}% of total return")
print(f"Dividend yield at purchase: {jnj['dividend_yield_at_purchase']}%")
print(f"Yield-on-cost today: {jnj['dividend_yield_on_cost_today']}%")
print()
print("This means the original $10,000 investment now generates:")
income = 10000 * (jnj['dividend_yield_on_cost_today'] / 100)
print(f"  ~${income:,.0f}/year in dividends alone")
```

## Key Takeaways

1. **DRIP compounds silently** — each dividend buys fractional shares that then generate their own dividends
2. **Yield-on-cost rises over time** — if the dividend grows and your cost basis is fixed, your effective yield grows
3. **DRIP matters most for high-yield, long holding periods** — Realty Income's DRIP contributed ~45% of total return
4. **DRIP is automatic compounding** — no decisions, no market timing, just reinvestment
