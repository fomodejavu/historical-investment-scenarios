# 📊 Real vs Nominal Returns: Inflation Adjustment

This notebook shows how inflation erodes the real value of returns, and compares nominal vs real (inflation-adjusted) gains across scenarios.

**Live calculator:** [fomodejavu.com](https://fomodejavu.com/)  
**Methodology:** [fomodejavu.com/methodology](https://fomodejavu.com/methodology/)

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Historical US CPI-U approximate annual inflation rates
cpi_annual = {
    1997: 2.3, 1998: 1.6, 1999: 2.2, 2000: 3.4, 2001: 2.8, 2002: 1.6,
    2003: 2.3, 2004: 2.7, 2005: 3.4, 2006: 3.2, 2007: 2.8, 2008: 3.8,
    2009: -0.4, 2010: 1.6, 2011: 3.2, 2012: 2.1, 2013: 1.5, 2014: 1.6,
    2015: 0.1, 2016: 1.3, 2017: 2.1, 2018: 2.4, 2019: 1.8, 2020: 1.2,
    2021: 7.0, 2022: 6.5, 2023: 3.4, 2024: 2.9
}

def cumulative_inflation(start_year: int, end_year: int) -> float:
    """Calculate cumulative inflation from start_year to end_year."""
    multiplier = 1.0
    for year in range(start_year, end_year):
        rate = cpi_annual.get(year, 2.5) / 100
        multiplier *= (1 + rate)
    return multiplier

def real_return(nominal_value: float, initial_investment: float,
                start_year: int, end_year: int) -> float:
    """Calculate real (inflation-adjusted) final value."""
    inflation_factor = cumulative_inflation(start_year, end_year)
    real_value = nominal_value / inflation_factor
    return real_value

# Example: Amazon IPO
amzn_nominal = 2050000
amzn_real = real_return(amzn_nominal, 1000, 1997, 2025)
inflation_factor = cumulative_inflation(1997, 2025)

print(f"Amazon IPO ($1,000 in 1997)")
print(f"  Nominal value today:          ${amzn_nominal:>15,.0f}")
print(f"  Cumulative inflation (1997-2025): {inflation_factor:.2f}×")
print(f"  Real value (2025 dollars):    ${amzn_real:>15,.0f}")
print(f"  Real still beats nominal entry by: {amzn_real/1000:.0f}×")
```

```python
# Compare nominal vs real returns across multiple scenarios
scenarios = [
    {'name': 'Amazon (1997)', 'nominal': 2050000, 'invested': 1000, 'start': 1997},
    {'name': 'Apple (2007)', 'nominal': 43000, 'invested': 1000, 'start': 2007},
    {'name': 'S&P 500 (2000 peak)', 'nominal': 5400, 'invested': 1000, 'start': 2000},
    {'name': 'S&P 500 (2009 bottom)', 'nominal': 9800, 'invested': 1000, 'start': 2009},
    {'name': 'VTI Total Market (2004)', 'nominal': 9500, 'invested': 1000, 'start': 2004},
    {'name': 'Cash (1990)', 'nominal': 1000, 'invested': 1000, 'start': 1990},
    {'name': 'Cash (2020)', 'nominal': 1000, 'invested': 1000, 'start': 2020},
]

for s in scenarios:
    s['real'] = real_return(s['nominal'], s['invested'], s['start'], 2025)
    s['nominal_return_pct'] = (s['nominal'] / s['invested'] - 1) * 100
    s['real_return_pct'] = (s['real'] / s['invested'] - 1) * 100

names = [s['name'] for s in scenarios]
nominal_vals = [s['nominal'] for s in scenarios]
real_vals = [s['real'] for s in scenarios]

x = range(len(names))
width = 0.38

fig, ax = plt.subplots(figsize=(13, 7))
bars1 = ax.bar([i - width/2 for i in x], nominal_vals, width, label='Nominal Value', color='#2563eb', alpha=0.8)
bars2 = ax.bar([i + width/2 for i in x], real_vals, width, label='Real Value (inflation-adjusted)', color='#10b981', alpha=0.8)
ax.axhline(1000, color='red', linestyle='--', alpha=0.5, label='$1,000 invested')

ax.set_xticks(list(x))
ax.set_xticklabels(names, rotation=22, ha='right', fontsize=9)
ax.set_ylabel('Value Today (USD, log scale)', fontsize=11)
ax.set_title('Nominal vs Real (Inflation-Adjusted) Returns\nAll scenarios assume $1,000 initial investment', fontsize=13)
ax.set_yscale('log')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}' if x < 1e6 else f'${x/1e6:.1f}M'))
ax.legend(fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

# Print summary table
print(f"{'Scenario':<28} {'Nominal':>12} {'Real':>12} {'Inflation Loss':>14}")
print('-' * 70)
for s in scenarios:
    loss = s['nominal'] - s['real']
    print(f"{s['name']:<28} ${s['nominal']:>10,.0f} ${s['real']:>10,.0f} ${loss:>12,.0f}")
```

## Key Insight

Even the most impressive nominal returns are reduced in real terms — but for strong equity investments, the reduction is marginal relative to the gain. **For cash, inflation is devastating over long periods.**
