# ₿ Crypto Volatility: Drawdown Analysis

This notebook illustrates the extreme volatility of crypto assets. The headline returns (e.g., Bitcoin 2012→2025) are real — but they required surviving multiple 80%+ crashes.

**⚠️ Volatility warning:** Crypto is one of the most volatile asset classes in financial history.

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import json

# Load Bitcoin 2012 scenario
with open('../scenarios/crypto/bitcoin-2012-1000usd.json') as f:
    btc = json.load(f)

print(f"Scenario: {btc['title']}")
print(f"Initial: ${btc['initial_investment_usd']:,}")
print(f"Value today: ${btc['value_today_usd']:,}")
print(f"Total return: {btc['total_return_pct']:,}%")
print(f"CAGR: {btc['cagr_pct']}%")
print()
print("Major drawdowns experienced:")
for dd in btc['major_drawdowns']:
    print(f"  {dd['peak_date']} → {dd['trough_date']}: {dd['drawdown_pct']}% drawdown")
    print(f"    ${dd['peak_price_approx']:,} → ${dd['trough_price_approx']:,}/BTC")
```

```python
# Simulate portfolio value through a hypothetical Bitcoin-like journey
# using approximate Bitcoin price history milestones

btc_price_history = [
    (2012.0,  5.27),
    (2013.9,  1163),
    (2015.1,  170),
    (2017.96, 19891),
    (2018.95, 3236),
    (2021.86, 68789),
    (2022.88, 15899),
    (2024.5,  70000),
    (2025.0,  97000),
]

initial_investment = 1000
btc_at_purchase = initial_investment / 5.27

years = [p[0] for p in btc_price_history]
portfolio_values = [btc_at_purchase * p[1] for p in btc_price_history]

fig, ax = plt.subplots(figsize=(13, 7))

ax.fill_between(years, portfolio_values, alpha=0.1, color='#f59e0b')
ax.plot(years, portfolio_values, color='#f59e0b', linewidth=2.5, label='Portfolio value ($1,000 initial)')
ax.axhline(initial_investment, color='red', linestyle='--', alpha=0.6, label='$1,000 initial investment')

# Annotate major crashes
ax.annotate('−85%\n($1.16M → $174K)', xy=(2015.1, 174*btc_at_purchase),
            xytext=(2016.5, 800000),
            arrowprops=dict(arrowstyle='->', color='#dc2626'),
            fontsize=9, color='#dc2626')
ax.annotate('−83%\n($20M → $3.3M)', xy=(2018.95, 3236*btc_at_purchase),
            xytext=(2019.5, 8000000),
            arrowprops=dict(arrowstyle='->', color='#dc2626'),
            fontsize=9, color='#dc2626')

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Portfolio Value (USD, log scale)', fontsize=12)
ax.set_title('$1,000 in Bitcoin (2012): The Volatility Journey\nThe headline return requires surviving multiple 80%+ crashes', fontsize=13)
ax.set_yscale('log')
ax.yaxis.set_major_formatter(plt.FuncFormatter(
    lambda x, _: f'${x:,.0f}' if x < 1e6 else f'${x/1e6:.1f}M'
))
ax.legend(fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

print("Key question: Could YOU have held through a 95% crash in 2015?")
print("After the 2013 peak: $1,000 investment was worth ~$1.16M, then fell to $174K")
print("Most investors sold. Those who held saw the subsequent recovery.")
```

## The Behavioral Finance Problem

The astronomical Bitcoin return is mathematically real — but behaviorally, almost no investor captured it fully because:

1. **Most sold during the 2013-2015 crash** (95% drawdown)
2. **Many bought near each peak** (not at the 2012 lows)
3. **Exchange hacks / lost wallets** destroyed holdings (Mt. Gox, 2014)
4. **FOMO buying at peaks** drove average investor returns far below buy-and-hold figures

These scenarios show what a perfect buy-and-hold investor would have earned. Real investor returns were dramatically lower.
